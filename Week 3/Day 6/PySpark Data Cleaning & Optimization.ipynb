```python
null_data = [
    (101, "Sravan", "IT", 75000),
    (102, None, "IT", 68000),
    (103, "Priya", None, 62000),
    (104, "Kiran", "HR", None),
    (105, None, None, None)
]

null_columns = ["emp_id", "name", "department", "salary"]

null_df = spark.createDataFrame(null_data, null_columns)

null_df.show()
```

```python

null_df.fillna({"salary": 0}).show()

null_df.fillna({"name": "Unknown"}).show()

null_df.fillna({"department": "NA"}).show()

null_df.fillna({"salary": 50000}).show()

null_df.fillna({
    "name": "Unknown",
    "department": "NA",
    "salary": 0
}).show()
```

```python

null_df.na.fill(0, ["salary"]).show()

null_df.na.fill("Unknown", ["name"]).show()

null_df.na.fill("IT", ["department"]).show()

null_df.na.fill(25).show()

null_df.na.fill({
    "name": "Unknown",
    "department": "NA",
    "salary": 0
}).show()
```

```python

null_df.na.drop().show()

null_df.na.drop(subset=["salary"]).show()

null_df.na.drop(subset=["department"]).show()

null_df.na.drop(thresh=3).show()

null_df.na.drop(subset=["name"]).show()
```

```python

df.replace("IT", "Technology", "department").show()

df.replace("HR", "Human Resources", "department").show()

df.replace("Hyderabad", "HYD", "city").show()

df.replace("Bangalore", "BLR", "city").show()

df.replace(
    ["IT", "HR"],
    ["Technology", "Human Resources"],
    "department"
).show()
```

```python
repartition_df = df.repartition(2)
repartition_df.show()

repartition_df = df.repartition(4)
repartition_df.show()

repartition_dept_df = df.repartition("department")
repartition_dept_df.show()

repartition_city_df = df.repartition("city")
repartition_city_df.show()

repartition_multi_df = df.repartition(
    "department",
    "city"
)
repartition_multi_df.show()
```
