```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("PySparkPractice").getOrCreate()

employee_data = [
    (101, "Sravan", "Data Engineer", "IT", 75000, "Hyderabad", 28, "2021-05-10", "Male"),
    (102, "Ravi", "Software Engineer", "IT", 68000, "Bangalore", 30, "2020-03-15", "Male"),
    (103, "Priya", "Data Analyst", "Analytics", 62000, "Chennai", 26, "2022-01-12", "Female"),
    (104, "Kiran", "Manager", "HR", 90000, "Mumbai", 35, "2018-07-19", "Male"),
    (105, "Anjali", "HR Executive", "HR", 45000, "Pune", 24, "2023-02-20", "Female"),
    (106, "Vikram", "Data Scientist", "Analytics", 98000, "Delhi", 32, "2019-11-25", "Male"),
    (107, "Sneha", "Developer", "IT", 71000, "Hyderabad", 27, "2021-08-17", "Female"),
    (108, "Rahul", "Tester", "QA", 55000, "Chennai", 29, "2020-06-10", "Male"),
    (109, "Meena", "QA Lead", "QA", 83000, "Bangalore", 33, "2017-09-14", "Female"),
    (110, "Arjun", "Support Engineer", "Support", 50000, "Pune", 31, "2022-04-11", "Male")
]

columns = ["emp_id","name","designation","department","salary","city","age","joining_date","gender"]

emp_df = spark.createDataFrame(employee_data, columns)

df = emp_df
```

```python
df.withColumn("designation_words",
              split(col("designation"), " ")).show()

df.withColumn("first_word",
              split(col("designation"), " ")[0]).show()

df.withColumn("word_count",
              size(split(col("designation"), " "))).show()
```

```python
df.withColumn(
    "emp_city",
    concat(col("name"), lit("_"), col("city"))
).show()

df.withColumn(
    "profile",
    concat(col("designation"), lit("-"), col("department"))
).show()
```

```python
df.withColumn("company_name", lit("Capgemini")).show()

df.withColumn("country", lit("India")).show()

df.withColumn("training_batch", lit("Data Engineering")).show()
```

```python
df.withColumn(
    "salary_category",
    when(col("salary") >= 90000, "High")
    .when(col("salary") >= 60000, "Medium")
    .otherwise("Low")
).show()

df.withColumn(
    "age_group",
    when(col("age") < 30, "Young")
    .otherwise("Experienced")
).show()
```

```python
df.withColumn(
    "name_prefix",
    substring(col("name"),1,3)
).show()

df.withColumn(
    "joining_year",
    substring(col("joining_date"),1,4)
).show()
```

```python
df.withColumn(
    "clean_name",
    regexp_replace(col("name")," ","")
).show()

df.withColumn(
    "updated_department",
    regexp_replace(col("department"),"IT","Technology")
).show()
```

```python
df.filter(col("name").like("S%")).show()

df.filter(col("city").like("%a%")).show()

df.filter(col("department").like("A%")).show()
```

```python
df.filter(
    col("city").isin("Hyderabad","Bangalore")
).show()

df.filter(
    col("department").isin("IT","HR")
).show()
```

```python
df.filter(
    col("salary").between(50000,80000)
).show()

df.filter(
    col("age").between(25,30)
).show()
```

```python
skill_df = df.withColumn(
    "skills",
    array(lit("Python"),lit("SQL"),lit("Spark"))
)

skill_df.select(
    "name",
    explode("skills").alias("skill")
).show()
```
