```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("PySparkPractice").getOrCreate()

employee_data = [
    (101, "Sravan", "Data Engineer", "IT", 75000, "Hyderabad", 28, "2021-05-10", "Male"),
    (102, "Ravi", "Software Engineer", "IT", 68000, "Bangalore", 30, "2020-03-15", "Male"),
    (103, "Priya", "Data Analyst", "Analytics", 62000, "Chennai", 26, "2022-01-12", "Female"),
    (104, "Kiran", "Manager", "HR", 90000, "Mumbai", 35, "2018-07-19", "Male"),
    (105, "Anjali", "HR Executive", "HR", 45000, "Pune", 24, "2023-02-20", "Female"),
    (106, "Vikram", "Data Scientist", "Analytics", 98000, "Delhi", 32, "2019-11-25", "Male"),
    (107, "Sneha", "Developer", "IT", 71000, "Hyderabad", 27, "2021-08-17", "Female"),
    (108, "Rahul", "Tester", "QA", 55000, "Chennai", 29, "2020-06-10", "Male"),
    (109, "Meena", "QA Lead", "QA", 83000, "Bangalore", 33, "2017-09-14", "Female"),
    (110, "Arjun", "Support Engineer", "Support", 50000, "Pune", 31, "2022-04-11", "Male")
]

columns = ["emp_id","name","designation","department","salary","city","age","joining_date","gender"]

emp_df = spark.createDataFrame(employee_data, columns)

df = emp_df
```

```python
df.drop("age").show()

df.drop("gender").show()

df.drop("joining_date").show()

df.drop("city","age").show()

df.drop("designation").show()
```

```python
df.select("department").distinct().show()

df.select("city").distinct().show()

df.select("designation").distinct().show()

df.select("gender").distinct().show()

df.select("department","city").distinct().show()
```

```python
df.dropDuplicates(["emp_id"]).show()

df.dropDuplicates(["department"]).show()

df.dropDuplicates(["city"]).show()

df.dropDuplicates(["department","city"]).show()

df.dropDuplicates(["name"]).show()
```

```python
df.groupBy("department").avg("salary").show()

df.groupBy("department").max("salary").show()

df.groupBy("department").min("age").show()

df.groupBy("city").count().show()

df.groupBy("gender").sum("salary").show()
```

```python
df.agg(sum("salary")).show()

df.agg(avg("age")).show()

df.agg(max("salary"), min("salary")).show()

df.agg(count("*")).show()

df.groupBy("department").agg(
    avg("salary").alias("avg_salary"),
    avg("age").alias("avg_age")
).show()
```

```python
department_data = [
("IT","John"),
("HR","Smith"),
("QA","David"),
("Analytics","Kevin"),
("Support","Robert")
]

dept_df = spark.createDataFrame(
    department_data,
    ["department","manager"]
)

emp_df.join(dept_df,"department","inner").show()

emp_df.join(dept_df,"department","left").show()

emp_df.join(dept_df,"department","right").show()

emp_df.join(dept_df,"department","outer").show()

emp_df.join(dept_df,"department") \
      .select("name","manager") \
      .show()
```

```python
new_emp_data = [
(111, "Teja", "Developer", "IT", 72000, "Hyderabad", 26, "2023-01-10", "Male"),
(112, "Divya", "Analyst", "Analytics", 65000, "Bangalore", 25, "2022-11-05", "Female")
]

new_emp_df = spark.createDataFrame(new_emp_data, columns)

df.union(new_emp_df).show()

df.union(new_emp_df).count()

df.union(new_emp_df).distinct().show()

df.union(new_emp_df) \
  .filter(col("department")=="IT") \
  .show()

df.union(new_emp_df) \
  .orderBy(col("salary").desc()) \
  .show()
```

```python
df.unionByName(new_emp_df).show()

df.unionByName(new_emp_df).count()

df.unionByName(new_emp_df).distinct().show()

df.unionByName(new_emp_df) \
  .filter(col("department")=="Analytics") \
  .show()

df.unionByName(new_emp_df) \
  .orderBy(col("salary").desc()) \
  .show()
```
